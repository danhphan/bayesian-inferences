```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma, factorial

%matplotlib inline
```

```python
math.gamma(2), math.gamma(0.5) ** 2
```

```python
gamma([0, 0.5, 1, 5])
```

```python
gamma(0.5)**2  # gamma(0.5) = sqrt(pi)
```

```python
x = np.linspace(-3.5, 5.5, 2051)
y = gamma(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'b', alpha=0.6, label='gamma(x)');
```

```python
plt.plot(x, y, 'b', alpha=0.6, label='gamma(x)');
plt.xlim(-3.5, 5.5)
plt.ylim(-10, 25)
```

```python
plt.plot(x, y, 'b', alpha=0.6, label='gamma(x)');
k = np.arange(1, 7)
plt.plot(k, factorial(k-1), 'k*', alpha=0.6,
         label='(x-1)!, x = 1, 2, ...')
plt.xlim(-3.5, 5.5)
plt.ylim(-10, 25)
```

```python
plt.plot(x, y, 'b', alpha=0.6, label='gamma(x)')
k = np.arange(1, 7)
plt.plot(k, factorial(k-1), 'k*', alpha=0.6,
         label='(x-1)!, x = 1, 2, ...')
plt.xlim(-3.5, 5.5)
plt.ylim(-10, 25)
plt.grid()
plt.xlabel('x')
plt.legend(loc='lower right')
plt.show()
```
